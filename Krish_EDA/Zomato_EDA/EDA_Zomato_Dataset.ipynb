```python
df_country = pd.read_excel(r"C:\Users\TUFAN\Downloads\res\ZOmato_EDA\Zomatodataset\Country-Code.xlsx")
```

```python
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

## Pie Chart
- Top 3 countries that uses zomato

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.show()
```

## Observations:
- Zomato maximum transaction are from India followed by US and UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observations:
- When rating is between 4.5 to 4.9 --> excellent
- When the ratings in between 4.0 to 4.4 --> very good
- When rating is between 3.5 to 3.9 --> Good
- When rating is between 3.0 to 3.4 --> Average
- When rating is between 2.5 to 2.9 --> Average
- When rating is between 1.8 to 2.4 --> Poor

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings,hue="Rating color",palette=['Blue','red','orange','yellow','green','green'])
plt.show()
```

## Observations:
- Not rated count is very high
- maximum number of ratings are between 2.5 to 3.4

## Count plot

```python
sns.countplot(x='Rating color',data=ratings,palette=['Blue','red','orange','yellow','green','green'])
```
